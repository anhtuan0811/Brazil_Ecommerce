```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

```

```python
orders = pd.read_csv('/content/olist_orders_dataset.csv')
payments = pd.read_csv('/content/olist_order_payments_dataset.csv')
```

```python
orders.head()
```

```python
orders.set_index('order_id', inplace=True)
orders.head()
```

```python
payments.set_index('order_id', inplace=True)
payments.head()
```

```python
payments.shape
```

```python
merge = orders.merge(payments, left_index=True, right_index=True, how='outer')
merge.head()
```

```python
merge.shape
```

```python
merge.isna().sum().sort_values(ascending=False)
```

```python
merge.dropna(inplace=True)
merge.isna().sum().sort_values(ascending=False)
```

```python
merge.dtypes
```

```python
merge[['order_purchase_timestamp',
       'order_approved_at',
       'order_delivered_carrier_date',
       'order_delivered_customer_date',
       'order_estimated_delivery_date']] = merge[['order_purchase_timestamp',
       'order_approved_at',
       'order_delivered_carrier_date',
       'order_delivered_customer_date',
       'order_estimated_delivery_date']].astype('datetime64[ns]')

merge[['payment_sequential', 'payment_installments']] = merge[['payment_sequential', 'payment_installments']].astype('int64')

merge.dtypes
```

```python
rfm = merge[['customer_id', 'order_purchase_timestamp', 'payment_value']].reset_index()
rfm.head()
```

```python
# Assume today is one day after the last purchase
today = rfm['order_purchase_timestamp'].max() + pd.to_timedelta(1, 'D')
today
```

```python
recency = pd.DataFrame(rfm.groupby(by='customer_id', as_index=False)['order_purchase_timestamp'].max())
recency.rename(columns={'order_purchase_timestamp': 'Last Purchase Date'}, inplace=True)
recency['Recency'] = recency['Last Purchase Date'].apply(lambda x: (today - x).days)
recency.head()
```

```python
frequency = pd.DataFrame(rfm.groupby(by='customer_id', as_index=False)['order_id'].count())
frequency.rename(columns={'order_id': 'Frequency'}, inplace=True)
frequency.head()
```

```python
monetary = pd.DataFrame(rfm.groupby(by='customer_id', as_index=False)['payment_value'].sum())
monetary.columns = ['customer_id', 'Monetary']
monetary
```

```python
df_rf = recency.merge(frequency, on='customer_id')
df_rfm = df_rf.merge(monetary, on='customer_id')
df_rfm.drop(columns='Last Purchase Date', inplace=True)
df_rfm
```

```python
df_rfm.describe().T
```

```python
def f_score(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    elif x == 3:
        return 3
    elif x == 4:
        return 4
    else:
        return 5



def m_score(x):
    if x < 50:
        return 1
    elif x < 100:
        return 2
    elif x < 150:
        return 3
    elif x < 200:
        return 4
    else:
        return 5

df_rfm['R Score'] = pd.qcut(x=df_rfm['Recency'], q=5, labels=[5, 4, 3, 2, 1])
df_rfm['F Score'] = df_rfm['Frequency'].apply(f_score )
df_rfm['M Score'] = x=df_rfm['Monetary'].apply(m_score)

df_rfm = df_rfm[['customer_id', 'Recency', 'R Score', 'Frequency', 'F Score', 'Monetary', 'M Score']]
df_rfm
```

```python
r = df_rfm['Recency']
r_score = df_rfm['R Score']

plt.scatter(r, r_score, s=100, c=r_score, edgecolor='black', linewidth=1, alpha=0.1)

cbar = plt.colorbar()
cbar.set_label('Recency Score')

plt.title('Recency Score Distribution')
plt.xlabel('Recency')
plt.ylabel('Recency Score')


plt.show()

# the lower the recency, the better
```

```python
f = df_rfm['Frequency']
f_score = df_rfm['F Score']

plt.scatter(f, f_score, s=100, c=f_score, edgecolor='black', linewidth=1, alpha=0.1)

cbar = plt.colorbar()
cbar.set_label('Frequency Score')

plt.title('Frequency Score Distribution')
plt.xlabel('Frequency')
plt.ylabel('Frequency Score')

plt.show()

```

```python
m = df_rfm['Monetary']
m_score = df_rfm['M Score']

plt.scatter(m, m_score, s=100, c=m_score, edgecolor='black', linewidth=1, alpha=0.1)

cbar = plt.colorbar()
cbar.set_label('Monetary Score')

plt.title('Monetary Score Distribution')
plt.xlabel('Monetary')
plt.ylabel('Monetary')

plt.show()

```

```python
df_rfm['R Score'] = df_rfm['R Score'].astype('int64')
df_rfm.dtypes
```

```python
df_rfm['RFM Score'] = df_rfm['R Score'].astype(str) + df_rfm['F Score'].astype(str) + df_rfm['M Score'].astype(str)
df_rfm.head()
```

```python
df_rfm['Segment'] = df_rfm['R Score'].astype(str) + df_rfm['F Score'].astype(str)
df_rfm.head()
```

```python
df_rfm['Segment'].unique()

# 5^3 = 125 possibilities at most
# so, for simplification, let's work with segments based on R Score and F Score only
```

```python
segt_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Cannot Loose',
    r'3[1-2]': 'About To Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}

df_rfm['Segment'] = df_rfm['Segment'].replace(segt_map, regex=True)
df_rfm.reset_index(inplace=True)
df_rfm.head()
```

```python
rfm_final = df_rfm[['customer_id', 'Recency', 'Frequency', 'Monetary', 'Segment']]
rfm_final.head()
```

```python
df_segt = rfm_final[['customer_id','Segment']].groupby('Segment').agg('count').reset_index()
df_segt.rename(columns={'customer_id': 'Customer Counts'}, inplace=True)
df_segt
```

```python
segment = df_segt['Segment']
customer_counts = df_segt['Customer Counts']

fig, ax = plt.subplots(figsize=(15,6))

plt.scatter(segment, customer_counts, s=customer_counts, c=customer_counts, edgecolor='black', linewidth=1, alpha=0.1)

cbar = plt.colorbar()
cbar.set_label('Customer Counts')

plt.title('Customer Counts By RFM Segmentation')
plt.ylabel('Customer Counts')

plt.tight_layout()
plt.show()
```
